# HTML etiketleri üzerinden Verileri çekme 

```python
import requests
from bs4 import BeautifulSoup
r = requests.get('https://github.com/')
print(r)
# parsing HTML -> HTML sayfası gibi yazdıralım
soup = BeautifulSoup(r.content, 'html.parser')
s = soup.find('div',class_ = 'container-xl p-responsive' )
baslik = s.find('h3', class_ ='h5 mt-4 mb-0')
print(baslik)
```

```python
import requests
from bs4 import BeautifulSoup
r = requests.get('https://github.com/')
print(r)
# parsing HTML -> HTML sayfası gibi yazdıralım
soup = BeautifulSoup(r.content, 'html.parser')
content = soup.find_all('p')
print(content)
```

web sitelerinin html olarak daha iyi bir şekilde görüntüleyebilmek için kendi Encoding desteğiyle kodlayarak açarsak bunu sağlayabiliriz.

```python
import urllib.request 
url ='https://www.example.com'

try:
    response =urllib.request.urlopen(url)
    data = response.read()

    #html kodunu görüntüleyebilmek için utf-8 (türkçe karakter) ile web sitesini görüntülüyoruz.
    html_content = data.decode('utf-8')
    print(html_content)
except Except as e:
    print('yanlış url', e)
```
