```python
import requests 
url = requests.get('https://www.nytimes.com/interactive/2017/06/23/opinion/trumps-lies.html')
```

```python
print(url.text[:500])
```

```python
from bs4 import BeautifulSoup
soup= BeautifulSoup(url.text, 'html.parser') # html etiketlerine kolayca ulaşmamızı sağlayacak.
print(soup.prettify())  #prettify() html blokları gözükür.
```

```python
results = soup.find_all('span', attrs={'class':'short-desc'}) #span classı short-desc olanları seçtik
```

```python
len(results) # 180 tane açıklama seçtik
```

# yukarıda belirlemiş olduğumuz yötemleri kullanarak veri setimizi yaratalım

```python
veriler =[]
for veri in results:
    date = veri.find('strong').text.replace('\xa0','')+', 2017'

    exp = veri.contents[1][1:-2]

    link = veri.find('a')['href']
    veriler.append((date,exp,link))
```

```python
len(veriler)
```

```python
veriler
```

```python
import pandas as pd
df = pd.DataFrame(data= veriler, columns=['Date', 'Expretion', 'Source'])
```
